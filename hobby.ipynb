```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('Hobby_Data.csv')
data.head()
```

```python
data.describe().T
```

```python
n_data = data.dropna()
```

```python
n_data["Predicted Hobby"] = n_data["Predicted Hobby"].map({"Academics": 0, "Sports": 1, "Arts": 2})
```

```python
# Loop through each predictor column and create bar plots
for col in n_data.columns[:-1]:  # Exclude 'Suicide attempt'
    groups = n_data[[col, 'Predicted Hobby']]
    grouped_data = groups.groupby(col)['Predicted Hobby'].mean()
    grouped_data = grouped_data.sort_values()

    # Plot
    fig, ax = plt.subplots(figsize=(16, 6), dpi=80)
    sns.barplot(x=grouped_data.index, y=grouped_data.values, ax=ax)
    ax.set_title(f"Mean Predicted Hobby Rate by '{col}' Response", fontsize=16)
    ax.set_ylabel("Mean Predicted Hobby Rate")
    ax.set_xlabel(col)
    ax.set_xticklabels(ax.get_xticklabels(), rotation=45)
    plt.tight_layout()
    plt.show()
```

```python
x = n_data.drop(columns=['Predicted Hobby'])
y = n_data['Predicted Hobby']
```

```python
from category_encoders import BinaryEncoder

categorical_features = ['Olympiad_Participation', 'Scholarship', 'School', 'Fav_sub', 'Projects','Grasp_pow', 'Time_sprt', 'Medals', 'Career_sprt', 'Act_sprt', 'Fant_arts', 'Won_arts', 'Time_art']

encoder = BinaryEncoder()
x_encoded = encoder.fit_transform(x[categorical_features])
x = x.drop(categorical_features, axis=1)
x = x.reset_index(drop=True)
x_encoded = x_encoded.reset_index(drop=True)
x = pd.concat([x, x_encoded], axis=1)
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.25, random_state=42)
```

```python
y.value_counts()
```

```python
from imblearn.over_sampling import SMOTE
smote = SMOTE()
x, y = smote.fit_resample(x, y)
```

```python
y.value_counts()
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.25, random_state=42)
```

```python
from sklearn.tree import DecisionTreeClassifier
dcs = DecisionTreeClassifier()
dcs.fit(x_train, y_train)
```

```python
from sklearn.metrics import accuracy_score, precision_score

dcs_prediction = dcs.predict(x_test)
accuracy = accuracy_score(y_test, dcs_prediction)   
precision = precision_score(y_test, dcs_prediction, average='macro')  # or 'weighted'

print(f'Accuracy: {accuracy * 100}')
print(f'Precision: {precision * 100}')
```

```python
from sklearn.tree import DecisionTreeClassifier, export_graphviz
import graphviz
from joblib import dump, load

dcs = DecisionTreeClassifier()
dcs.fit(x_train, y_train)
export_graphviz(dcs, out_file='tree.dot', filled=True, feature_names=x_train.columns)

with open('tree.dot') as f:
    dot_graph = f.read()
graphviz.Source(dot_graph)

dump(dcs, 'decision_tree_model.joblib')
```
